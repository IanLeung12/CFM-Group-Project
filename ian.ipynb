```python
df = pd.DataFrame(columns=["Ticker", "Industry", "Score"])
industries = ["A", "B", "C", "D"]
for i in range(40):
    df.loc[i] = [chr(ord('a') + i), random.choice(industries), random.randint(10, 1000)]

df = df.sort_values(by="Score", ascending=False).reset_index(drop=True)
print(df)
```

```python
def find_percent(size):
    if size < 4: 
        percent = 15
        size += 1
    elif size < 6:
        percent = 10
        size += 1
    else:
        percent = 5
        size += 1
    return percent
```

```python
portfolio = {}
industry_set = {}
size = 0
for i, row in df.iterrows():
    if row["Industry"] not in industry_set:
        percent = find_percent(size)
        size += 1
        portfolio[row["Ticker"]] = percent
        industry_set[row["Industry"]] = percent
    else:
        percent = find_percent(size)
        percent = min(40 - industry_set[row["Industry"]], percent)
        if percent <= 0:
            continue
        size += 1
        portfolio[row["Ticker"]] = percent
        industry_set[row["Industry"]] += percent
    if size >= 10:
        break
print(portfolio)    
print(industry_set)
        

```
